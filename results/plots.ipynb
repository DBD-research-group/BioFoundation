## <u>Plots & Graphs for visualising the results</u>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Heatmap

```python
def create_hm(df,title):
    # Create the heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(df, annot=True, cmap='Blues', fmt='.2f') #Good: RdBu, Viridis, Blues

    # Add labels and title
    plt.title(title)
    plt.xlabel('Foundation Model')
    plt.ylabel('Animal dataset')
    return plt
    
```

### Linear probing

```python
df = pd.read_csv('probing.csv',  index_col='Dataset/Model')

plt = create_hm(df, 'Linear Probing accuracy of different Foundation Models')

# Save the heatmap
plt.savefig('probing_hm.png', dpi=300, bbox_inches='tight')

# Show the heatmap
plt.show()
```

### Finetuning

```python
df = pd.read_csv('finetuning.csv',  index_col='Dataset/Model')

plt = create_hm(df, 'Finetuning accuracy of different Foundation Models')

# Save the heatmap
plt.savefig('finetuning_hm.png', dpi=300, bbox_inches='tight')

# Show the heatmap
plt.show()
```

```python
# Read the finetuning and probing CSV files
df_finetuning = pd.read_csv('finetuning.csv', index_col='Dataset/Model')
df_probing = pd.read_csv('probing.csv', index_col='Dataset/Model')

# Calculate the difference between the two DataFrames
df_difference = df_finetuning - df_probing

# Calculate the mean of the difference
mean_difference = df_difference.mean()

mean_difference
```

## Other plots

```python
# Data
models = ['EAT', 'SSAST', 'BEATs', 'AVES']
params = [25.5, 89, 90, 95]  # Params in millions
accuracy = [96, 88.8, 98, 77]  # Accuracy in percentage

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(params, accuracy, marker='o', linestyle='-', color='turquoise', label="Models")

# Add labels for each point
for i, model in enumerate(models):
    plt.text(params[i], accuracy[i], f"{model}", fontsize=10, ha='right')

# Add title and labels
plt.title('Model Accuracy vs Parameters')
plt.xlabel('Parameters (Millions)')
plt.ylabel('Top-1 Accuracy (%)')

# Customize axes
plt.grid(True)
plt.xticks(params)
plt.yticks(range(70, 101, 5))  # Adjust y-axis to show 70-100 accuracy

# Display the plot
plt.show()
```
