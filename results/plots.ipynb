## <u>Plots & Graphs for visualising the results</u>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Heatmap

```python
df = pd.read_csv('probing.csv',  index_col='Dataset/Model')

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='viridis', fmt='.2f')

# Add labels and title
plt.title('Accuracy Heatmap of different Foundation Models')
plt.xlabel('Foundation Model')
plt.ylabel('Animal Dataset')

# Show the heatmap
plt.show()
```
