### Create Dataframe from all files

```python
# Initialize an empty dataframe
df = pd.DataFrame()

# Loop through the files in the LABELS directory
for file_name in os.listdir(LABELS):
    if file_name.endswith(".csv"):
        file_path = os.path.join(LABELS, file_name)
        temp_df = pd.read_csv(file_path)
        df = pd.concat([df, temp_df], ignore_index=True)

# Display the dataframe
df.head()
```

## Total length of recordings

```python
import soundfile as sf

total_length = 0

# Loop through the files in the AUDIOS directory
for file_name in os.listdir(AUDIOS):
    if file_name.endswith(".flac"):
        file_path = os.path.join(AUDIOS, file_name)
        with sf.SoundFile(file_path) as audio_file:
            total_length += len(audio_file) / audio_file.samplerate

print(f"Total length of all audio files: {total_length / 3600:.2f} hours")

# Check how this compares to annotated duration
df["duration"] = df["end_time"] - df["start_time"]
total_duration_from_df = df["duration"].sum()

print(
    f"Total length of all audio files from dataframe (overlap possible): {total_duration_from_df / 3600:.2f} hours"
)
```

## Get Metadata for all 59 birds of the O4 Test dataset from XCL

```python
# download XCL metadata from huggingface
from huggingface_hub import hf_hub_download

# Download the parquet file from the dataset repo
file_path = hf_hub_download(
    repo_id="DBD-research-group/BirdSet",
    repo_type="dataset",
    filename="XCL/XCL_metadata.parquet",
    revision="data",
)

import pandas as pd

xcl = pd.read_parquet(file_path)
xcl.head()
# print all columns in the dataframe
print("Columns in the XCL metadata dataframe:")
for column in xcl.columns:
    print(f"- {column}")
```

```python
# filter the XCL metadata to only include rows with an ebird_code that is in the oekofor dataframe
xcl_filtered = xcl[xcl["ebird_code"].isin(oekofor["ebird_code"])]
# print the number of rows in the filtered dataframe
print(f"Number of rows in the filtered XCL metadata dataframe: {len(xcl_filtered)}")
```

### Samples per species

```python
SPECIES = xcl_filtered["ebird_code"].value_counts().to_dict()
print("Samples per ebird code:")
print(SPECIES)

unique_common_names = len(SPECIES.keys())
unique_ebird = xcl_filtered["ebird_code"].unique()

print(f"Number of ebird codes: {len(unique_ebird)}")
```

### Visualize the filtered metadata

```python
# Plot the common names and their counts
plt.figure(figsize=(10, 8))
plt.barh(list(SPECIES.keys()), list(SPECIES.values()))
plt.xlabel("Count")
plt.ylabel("Common Name")
plt.title("Common Name Counts")
plt.show()
```

### Plot the mean and variance of the length of the recordings per species

```python
# Calculate mean and variance of 'length' per species (ebird_code)
mean_length = xcl_filtered.groupby("ebird_code")["length"].mean()
var_length = xcl_filtered.groupby("ebird_code")["length"].var()

# Create a dataframe for plotting
length_stats = pd.DataFrame({"mean_length": mean_length, "var_length": var_length})

# Plot mean and variance
fig, ax1 = plt.subplots(figsize=(14, 8))

color = "tab:blue"
ax1.set_xlabel("ebird_code")
ax1.set_ylabel("Mean Length (s)", color=color)
ax1.bar(
    length_stats.index,
    length_stats["mean_length"],
    color=color,
    alpha=0.6,
    label="Mean Length",
)
ax1.tick_params(axis="y", labelcolor=color)
ax1.set_xticklabels(length_stats.index, rotation=90)

ax2 = ax1.twinx()
color = "tab:red"
ax2.set_ylabel("Variance of Length", color=color)
ax2.plot(
    length_stats.index,
    length_stats["var_length"],
    color=color,
    marker="o",
    linestyle="-",
    label="Variance",
)
ax2.tick_params(axis="y", labelcolor=color)

plt.title("Mean and Variance of Recording Length per Species")
fig.tight_layout()
plt.show()
```

## Sample 10 recordings per specie

Use recordings that are as close as possible to a length of 5s.

```python
sampled_recordings = []

for ebird_code in xcl_filtered["ebird_code"].unique():
    species_df = xcl_filtered[xcl_filtered["ebird_code"] == ebird_code].copy()
    # Compute absolute difference from 5 seconds
    species_df["length_diff"] = (species_df["length"] - 5).abs()
    # Sort by closeness to 5s and take 10 closest
    closest = species_df.nsmallest(10, "length_diff")
    sampled_recordings.append(closest)

# Concatenate all sampled recordings into a single dataframe
sampled_df = pd.concat(sampled_recordings, ignore_index=True)

# Drop the helper column
sampled_df = sampled_df.drop(columns=["length_diff"])

# Show the result
sampled_df[["ebird_code", "filepath", "length"]].head(20)
```

```python
# export the sampled recordings to a CSV file
sampled_df.to_csv("XCL_Sampled_Recordings_Batch_" "0.csv", index=False)
```
