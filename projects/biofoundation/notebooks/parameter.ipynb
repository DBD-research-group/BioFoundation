```python
import matplotlib.pyplot as plt
import numpy as np

# colors
blue = '#1f77b4'
orange = '#ff7f0e'
green = '#2ca02c'
red = '#d62728'

# size
outer_sizes = [90.3, 1.2 + 0.0161]
inner_sizes = [1.2, 0.0161]

# Text
outer_labels = ['Encoder Layer\n90.3M', 'Training Layers\n1.22M']
inner_labels = ['Attentive Pooling\n1.2M', 'Linear Probing\n16.1K']

# colors
outer_colors = [blue, orange]
inner_colors = [green, red]

fig, ax = plt.subplots(figsize=(8, 8))

# outside layer(Encoder و Training total)
outer_wedges, _ = ax.pie(
    outer_sizes, radius=1, colors=outer_colors,
    wedgeprops=dict(width=0.3, edgecolor='white'),
    startangle=90
)

# inside layer (Attentive و Linear)
inner_wedges, _ = ax.pie(
    inner_sizes, radius=0.7, colors=inner_colors,
    wedgeprops=dict(width=0.3, edgecolor='white'),
    startangle=90
)

# add labels to outer wedges

def add_inner_label(wedge, text, radius, text_color='white'):
    angle = (wedge.theta2 + wedge.theta1) / 2
    x = radius * np.cos(np.deg2rad(angle))
    y = radius * np.sin(np.deg2rad(angle))
    ax.text(x, y, text, ha='center', va='center', fontsize=9, color=text_color, weight='bold')

# add labels to outer and inner wedges

add_inner_label(outer_wedges[0], outer_labels[0], 0.85)  # 
add_inner_label(outer_wedges[1], outer_labels[1], 0.85, text_color='black')  #
add_inner_label(inner_wedges[0], inner_labels[0], 0.55)  # 
add_inner_label(inner_wedges[1], inner_labels[1], 0.55, text_color='black')  # 

# title
#plt.title('Parameter Distribution: Feature Extractor vs Trainable Layers', fontsize=13)
plt.axis('equal')
plt.tight_layout()
plt.savefig("param_distribution_all_inside.pdf", format="pdf")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
encoder_params = 90.3
attentive_pooling_params = 1.2
linear_probing_params = 0.0161  # = 16.1K

# 
sizes = [encoder_params, attentive_pooling_params, linear_probing_params]
colors = ['#ff7f0e', '#2ca02c', '#d62728']  # 
labels = ['Encoder\n(Frozen)\n90.3M', 'Attentive Pooling\n1.2M', 'Linear Probing\n16.1K']
text_colors = ['white', 'white', 'white']

# 
radii = [0.6, 0.6, 0.45]

# 
font_sizes = [10, 9, 8]

# 
fig, ax = plt.subplots(figsize=(8, 8))
wedges, _ = ax.pie(sizes, colors=colors, startangle=140)

# 
for i, wedge in enumerate(wedges):
    angle = (wedge.theta2 + wedge.theta1) / 2
    x = radii[i] * np.cos(np.deg2rad(angle))
    y = radii[i] * np.sin(np.deg2rad(angle))
    ax.text(x, y, labels[i], ha='center', va='center',
            fontsize=font_sizes[i], color=text_colors[i], weight='bold')

# title

#plt.title('Parameter Distribution: Feature Extractor vs Trainable Layers', fontsize=13)

# 
plt.axis('equal')
plt.tight_layout()

# 
plt.savefig("pie_all_inside.pdf", format="pdf")

#
plt.show()
```
