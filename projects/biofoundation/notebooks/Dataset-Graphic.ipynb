## Datasets Graphic
In  this notebook we visualize the size of different taxonomic groups of Big libraries:"XC, MAC, INA, ASA". 
We consider these toxonomic groups: " Birds/ Mammals(Mammals+Bats)/ Amphibians/ Insects/ Reptiels"

### Real Numbers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Example data: Replace this with your actual values
data = {
    "XC": [873376, 2486, 4098, 32082, 0],   # XC recordings for each category
    "MAC": [2669609, 11542, 9515, 9060, 63], # MAC recordings
    "INS": [871771, 94874, 47631, 80545, 770], # INS recordings
    "ASA": [21285, 692, 2716, 738, 1]   # ASA recordings
}

categories = ["Birds", "Amphibians", "Mammals", "Insects", "Reptiles"]

df = pd.DataFrame(data, index=categories)

# Define colors for different datasets
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(8, 5))
df.plot(kind="bar", stacked=True, color=colors, ax=ax)

# Labels and title
ax.set_ylabel("Number of Recordings")
ax.set_xlabel("Categories")
ax.set_title("Stacked Bar Chart of Recordings by Dataset")

# Legend outside
ax.legend(title="Datasets", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

## Seperated Charts

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Example data: Replace this with your actual values
data = {
    "XC": [873376, 2486, 4098, 32082, 0],   # XC recordings for each category
    "MAC": [2669609, 11542, 9515, 9060, 63], # MAC recordings
    "INS": [871771, 94874, 47631, 80545, 770], # INS recordings
    "ASA": [21285, 692, 2716, 738, 1]   # ASA recordings
}

categories = ["Birds", "Amphibians", "Mammals", "Insects", "Reptiles"]
df = pd.DataFrame(data, index=categories)

# Define colors for different datasets
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]

# Set bar width
bar_width = 0.2  
x = np.arange(len(categories))  # X locations for categories

# Plot each dataset as separate bars
fig, ax = plt.subplots(figsize=(8, 5))
for i, (dataset, color) in enumerate(zip(df.columns, colors)):
    ax.bar(x + i * bar_width, df[dataset], width=bar_width, label=dataset, color=color)


# Labels and title
ax.set_ylabel("Number of Recordings")
ax.set_xlabel("Categories")
ax.set_title("Grouped Bar Chart of Recordings by Dataset")

# X-axis settings
ax.set_xticks(x + bar_width * (len(df.columns) / 2 - 0.5))
ax.set_xticklabels(categories, rotation=45)

# Legend
```

## Logarithmic Scale

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Real data
data = {
    "XC": [873376, 2486, 4098, 32082, 0],   
    "MAC": [2669609, 11542, 9515, 9060, 63], 
    "INS": [871771, 94874, 47631, 80545, 770], 
    "ASA": [21285, 692, 2716, 738, 1]   
}

categories = ["Birds", "Amphibians", "Mammals", "Insects", "Reptiles"]
df = pd.DataFrame(data, index=categories)

# Define colors for different datasets
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(8, 5))
df.plot(kind="bar", stacked=True, color=colors, ax=ax)

# Apply log scale to y-axis
ax.set_yscale("log")

# Labels and title
ax.set_ylabel("Number of Recordings (log scale)")
ax.set_xlabel("Categories")
ax.set_title("Stacked Bar Chart of Recordings by Dataset (Log Scale)")

# Legend
ax.legend(title="Datasets", bbox_to_anchor=(1.05, 1), loc="upper left")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)  
plt.show()
```

## Normalize Data (0–100%)

```python
# Normalize each category to 100% (percentage-based)
df_percent = df.div(df.sum(axis=1), axis=0) * 100

# Plot normalized stacked bar chart
fig, ax = plt.subplots(figsize=(8, 5))
df_percent.plot(kind="bar", stacked=True, color=colors, ax=ax)

# Labels and title
ax.set_ylabel("Percentage of Recordings (%)")
ax.set_xlabel("Categories")
ax.set_title("Stacked Bar Chart of Recordings by Dataset (Normalized)")

# Legend outside
ax.legend(title="Datasets", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.xticks(rotation=45)  
plt.show()
```
